```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
sns.distplot(df["bill_length_mm"], kde=False, bins=7)
```

```python
sns.distplot(df["bill_length_mm"],kde=True,bins=7)
```

```python
sns.jointplot(x="bill_length_mm",y="body_mass_g",data=df,kind="hex")
```

```python
sns.jointplot(x="bill_length_mm",y="body_mass_g",data=df,hue="species",palette="Set2")
```

```python
sns.jointplot(x="bill_length_mm",y="body_mass_g",kind="hist",data=df)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,palette="husl",hue="species")
```
