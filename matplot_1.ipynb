1. Load the necessary package for plotting using pyplot from **matplotlib**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
Days=[1,2,3,4,5,6,7,8,9]
Speed=[78,45,67,23,78,45,34,57,77]
plt.plot(Days,Speed)
plt.title("car speed measurment")
plt.xlabel('days')
plt.ylabel("car speed")
plt.show()
```

2. Now to above car data apply some string formats like line style example green
dotted line, marker shape like +, change markersize, markerface color etc.\






```python
Days=[1,2,3,4,5,6,7,8,9]
Speed=[78,45,67,23,78,45,34,57,77]
plt.plot(Days,Speed, "--", marker="*", color="g", label="CSD")
plt.scatter(Days,Speed, c="b", s=45)
plt.title(" car speed measurment")
plt.legend()
plt.xlabel("days")
plt.ylabel("car speed")
plt.show()
```

3. Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average
speed in 8 days

```python
Days=[1,2,3,4,5,6,7,8,9]
max_speed=[78,75,89,78,90,92,89,68,99]
min_speed=[12,34,32,40,27,15,45,10,22]
avg_speed=[45,55,61,59,58,53,67,39,60]
plt.plot(Days,max_speed,"r",label="maximum speed")
plt.plot(Days,min_speed,"g",label="minimum speed")
plt.plot(Days,avg_speed,"b",label="average speed")
plt.title("car speed measurment")
plt.xlabel("Days")
plt.ylabel("car speed")
plt.legend()
plt.grid()
plt.show()
```

4. Plotting a basic sine graph by adding more features. Adding Multiple
plots by Superimposition like cosine wave

```python
x=np.arange(0,6*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,x,z)
plt.show()
```

5. Plot Simple bar chart showing popularity of Programming

```python
Languages=['python','sql','java','c++','javascript']
Popularity=[78,60,85,90,65]
Security=[44,36,55,50,42]
plt.bar(Languages,Popularity,color='b',label='Popularity',width=0.8)
plt.bar(Languages,Security,color='y',label='Security',width=0.4)
plt.legend()
plt.show()
```

```python
plt.barh(Languages,Popularity,color='g',label='Popularity')
plt.legend()
plt.show()
```

6. Plot Histogram, We have a sample data of Students marks of various Students, we will
try to plot number of Students by marks range and try to fgure out how many Students
are average, below-average and Excellent

```python
marks=np.array([67,45,78,90,34,58,89,92,72,23,45,66,77,34,89,98])
below_avg=marks[np.logical_and(marks>=40, marks<60)]
average=marks[np.logical_and(marks>=60,marks<80)]
excellent=marks[np.logical_and(marks>=80,marks<100)]
plt.hist(marks,color='m')
plt.title('marks chart')
plt.xlabel('bins')
plt.ylabel("marks")
plt.show()
```
