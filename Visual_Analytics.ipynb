```python
 import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the arrays
arr1 = np.array([23, 56, 66, 77, 89, 90])
arr2 = np.array([90, 56, 78, 34, 56, 45])

# Line plot
plt.figure(figsize=(8, 6))
plt.plot(arr1,arr2)
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(arr1,arr2)
plt.title('Bar Plot')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(arr1, arr2)
plt.title('Scatter Plot')
plt.xlabel('Array 1')
plt.ylabel('Array 2')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(arr1, labels=range(len(arr1)), autopct='%1.1f%%')
plt.title('Pie Plot - Array 1')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(arr1)
plt.title('Histogram - Array 1')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(arr2)
plt.title('Histogram - Array 2')
plt.show()
```

```python
file=r"/titanicdataset.csv"
titanic_df = pd.read_csv(file)



# Convert the 'survived' column to numeric values (0 and 1)
titanic_df['Survived'] = titanic_df['Survived'].astype(int)

# Convert the 'Age' column to categorical age groups
age_bins = [0, 18, 30, 50, 100]
age_labels = ['0-17', '18-29', '30-49', '50+']
titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'], bins=age_bins, labels=age_labels)

# Group by AgeGroup and calculate the survival rate
survival_rates = titanic_df.groupby('AgeGroup')['Survived'].mean()

# Plot the line plot
plt.figure(figsize=(8, 5))
plt.plot(survival_rates.index, survival_rates.values, marker='o')
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(survival_rates.index, survival_rates.values)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(survival_rates.index, survival_rates.values, marker='o')
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
