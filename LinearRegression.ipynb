```python
# Data Visualization
plt.scatter(df.Hours,df.Scores,marker="*",color="r")
plt.title("Student Score VS Hours", weight="bold")
plt.xlabel('Hours')
plt.ylabel("Scores")
plt.show()
```

```python
x=np.array(df.Hours)
y=np.array(df.Scores)
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=.3)
print(x_train)
print("Shape : ", x_train.shape)
```

```python
print(x_test)
print("Shape : ",x_test.shape)
```

```python
print(y_train)
print("Shape : ",y_train.shape)
```

```python
print(y_test)
print("Shape : ",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg_model=LinearRegression()
lin_reg_model.fit(x_train, y_train)
print(" model is ready for prediction")
```

```python
print(" Coefficient (a) : ", lin_reg_model.coef_)
print(" Intercept (b) : ", lin_reg_model.intercept_)
```

```python
Y= lin_reg_model.coef_*x_train + lin_reg_model.intercept_

plt.scatter(x_train,y_train,color="b",marker="+")
plt.plot(x_train,Y,color="g")
plt.title(" Regression line ( Best Fit Line) for Training Data", weight='bold')
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
y_predic = lin_reg_model.predict(x_test)
print(y_predic)
```

```python
y_test
```

```python
plt.scatter(x_test,y_test,color="b",marker="+")
plt.plot(x_test,y_predic,color="g")
plt.title(" Regression line (Best Fit Line) for Testing Data",weight = 'bold')
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```
